```python
import os
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
try:
    from xgboost import XGBRegressor
    HAS_XGB = True
except Exception:
    HAS_XGB = False
```

```python
FILE_PATH = r"C:\Users\samru\Desktop\Portfolio\Projects\Python\Demand Forecast\demand_forecasting_dataset.csv"

if not Path(FILE_PATH).exists():
    raise FileNotFoundError(f"CSV not found at {FILE_PATH}. Please confirm the path.")

df = pd.read_csv(FILE_PATH)
```

```python
print('Rows, cols:', df.shape)
print(df.columns.tolist())
```

```python
if 'date' not in df.columns:
    raise KeyError("The dataset must contain a 'date' column")
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date']).sort_values('date').reset_index(drop=True)
df = df.set_index('date')
```

```python
display(df.head())
print(df.info())
print(df.describe().T)
```

```python
# Feature engineering
df['year'] = df.index.year
df['month'] = df.index.month
df['day'] = df.index.day
df['day_of_week'] = df.index.dayofweek
df['week_of_year'] = df.index.isocalendar().week.astype(int)
```

```python
# lag / rolling on sales_units
if 'sales_units' not in df.columns:
    raise KeyError("The dataset must contain 'sales_units' column for lag features")


df['lag_1'] = df['sales_units'].shift(1)
df['lag_7'] = df['sales_units'].shift(7)
df['lag_30'] = df['sales_units'].shift(30)


df['rolling_7'] = df['sales_units'].rolling(window=7).mean()
df['rolling_30'] = df['sales_units'].rolling(window=30).mean()
```

```python
df = df.dropna()
print('After features, shape:', df.shape)
```

```python
# Train-test split (time-based)
TEST_DAYS = 365
feature_cols = [c for c in df.columns if c != 'future_demand']
X = df[feature_cols]
y = df['future_demand']


train_size = len(df) - TEST_DAYS
X_train = X.iloc[:train_size]
X_test = X.iloc[train_size:]
y_train = y.iloc[:train_size]
y_test = y.iloc[train_size:]


print('Train shape:', X_train.shape, 'Test shape:', X_test.shape)
```

```python
from pathlib import Path

PLOTS_DIR = Path('plots')
RESULTS_DIR = Path('results')


PLOTS_DIR.mkdir(exist_ok=True, parents=True)
RESULTS_DIR.mkdir(exist_ok=True, parents=True)
```

```python
# Train RandomForest
rf = RandomForestRegressor(n_estimators=300, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)


rf_mae = mean_absolute_error(y_test, rf_preds)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_preds))
print('RandomForest MAE:', rf_mae, 'RMSE:', rf_rmse)


# save RF predictions
results = df.iloc[train_size:].copy()
results['predicted_rf'] = rf_preds
results.to_csv(RESULTS_DIR / 'results_with_rf_predictions.csv')
```

```python
# plot actual vs predicted
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(y_test.values, label='Actual')
ax.plot(rf_preds, label='RF Predicted')
ax.set_title('Actual vs RF Predicted (test period)')
ax.legend()
fig.savefig(PLOTS_DIR / 'actual_vs_rf_predicted.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Feature importance (RF)
feat_imp = pd.Series(rf.feature_importances_, index=X_train.columns).sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(10,6))
feat_imp.head(15).plot(kind='bar', ax=ax)
ax.set_title('Top 15 Feature Importances (RF)')
fig.savefig(PLOTS_DIR / 'feature_importances_rf.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Train XGBoost
if HAS_XGB:
    xgb = XGBRegressor(
n_estimators=500,
learning_rate=0.05,
max_depth=6,
subsample=0.8,
colsample_bytree=0.8,
random_state=42
)
xgb.fit(X_train, y_train)
xgb_preds = xgb.predict(X_test)
xgb_mae = mean_absolute_error(y_test, xgb_preds)
xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_preds))
print('XGBoost MAE:', xgb_mae, 'RMSE:', xgb_rmse)
```

```python
if HAS_XGB:
    fig, ax = plt.subplots(figsize=(12,5))
    ax.plot(y_test.values, label='Actual')
    ax.plot(xgb_preds, label='XGB Predicted')
    ax.set_title('Actual vs XGB Predicted (test period)')
    ax.legend()
    fig.savefig(PLOTS_DIR / 'actual_vs_xgb_predicted.png', dpi=300, bbox_inches='tight')
    plt.show()
else:
    print('XGBoost not installed; skipping XGBoost model.')
```

```python
def forecast_future(model, df_history, days, feature_cols):
    df_future = df_history.copy()

    for _ in range(days):
        next_date = df_future.index.max() + pd.Timedelta(days=1)
        last_row = df_future.iloc[-1]
        new_row = {}

        # Date features
        new_row['year'] = next_date.year
        new_row['month'] = next_date.month
        new_row['day'] = next_date.day
        new_row['day_of_week'] = next_date.dayofweek
        new_row['week_of_year'] = next_date.isocalendar().week

        # Copy static features
        for col in feature_cols:
            if col not in new_row:
                new_row[col] = last_row[col] if col in last_row.index else np.nan

        # Lag features
        new_row['lag_1'] = last_row['future_demand']
        new_row['lag_7'] = (
            df_future['future_demand'].iloc[-7]
            if len(df_future) >= 7 else last_row['future_demand']
        )
        new_row['lag_30'] = (
            df_future['future_demand'].iloc[-30]
            if len(df_future) >= 30 else last_row['future_demand']
        )

        # Rolling window features
        new_row['rolling_7'] = df_future['future_demand'].iloc[-7:].mean()
        new_row['rolling_30'] = df_future['future_demand'].iloc[-30:].mean()

        # Convert to DataFrame, reorder columns
        new_X = pd.DataFrame([new_row], index=[next_date])
        new_X = new_X.reindex(columns=feature_cols)

        # Predict
        pred = model.predict(new_X)[0]
        new_X['future_demand'] = pred

        # Append
        df_future = pd.concat([df_future, new_X])

    return df_future.tail(days)[['future_demand']]
```

```python
# Forecast 30 and 365 days using RF and XGB 
feature_cols = [c for c in df.columns if c != 'future_demand']
forecast_30_rf = forecast_future(rf, df, 30, feature_cols)
forecast_30_rf.to_csv(RESULTS_DIR / 'forecast_30_rf.csv')


if HAS_XGB:
    forecast_30_xgb = forecast_future(xgb, df, 30, feature_cols)
forecast_30_xgb.to_csv(RESULTS_DIR / 'forecast_30_xgb.csv')


forecast_365_rf = forecast_future(rf, df, 365, feature_cols)
forecast_365_rf.to_csv(RESULTS_DIR / 'forecast_365_rf.csv')
if HAS_XGB:
    forecast_365_xgb = forecast_future(xgb, df, 365, feature_cols)
forecast_365_xgb.to_csv(RESULTS_DIR / 'forecast_365_xgb.csv')
```

```python
# Plot forecasts (30-day)
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(df.index[-200:], df['future_demand'].iloc[-200:], label='Historical (recent)')
ax.plot(forecast_30_rf.index, forecast_30_rf['future_demand'], marker='o', label='RF 30-day')
if HAS_XGB:
    ax.plot(forecast_30_xgb.index, forecast_30_xgb['future_demand'], marker='x', label='XGB 30-day')
ax.set_title('30-day Forecast vs Recent Historical')
ax.legend()
fig.savefig(PLOTS_DIR / 'forecast_30_day_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Combine historical + forecasts 
combined = pd.concat([
df[['future_demand']].rename(columns={'future_demand': 'Historical Demand'}),
forecast_30_rf.rename(columns={'future_demand': 'RF 30-day Forecast'})
], axis=0)
if HAS_XGB:
    combined = pd.concat([combined, forecast_30_xgb.rename(columns={'future_demand': 'XGB 30-day Forecast'})], axis=0)


combined = combined.sort_index()
fig, ax = plt.subplots(figsize=(14,6))
ax.plot(combined.index, combined['Historical Demand'], label='Historical Demand')
if 'RF 30-day Forecast' in combined.columns:
    ax.plot(combined.index, combined['RF 30-day Forecast'], label='RF 30-day Forecast')
if 'XGB 30-day Forecast' in combined.columns:
    ax.plot(combined.index, combined['XGB 30-day Forecast'], label='XGB 30-day Forecast')
ax.set_xlim(pd.to_datetime('2036-01-01'), combined.index.max())
ax.set_title('Historical Demand + 30-Day Forecast (Zoomed)')
ax.legend()
fig.savefig(PLOTS_DIR / 'historical_plus_30day_forecast.png', dpi=300, bbox_inches='tight')
plt.show()
```
